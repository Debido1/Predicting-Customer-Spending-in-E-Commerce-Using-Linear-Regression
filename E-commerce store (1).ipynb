### SPLITTING INTO FEATURES AND TARGET

```python
x = df_copy[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = df_copy[['Yearly Amount Spent']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
df_copy
```

### LINEAR REGRESSION MODEL TRAINNING AND PREDICTION

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

I have successfully created the model and its time to train the model with our dataset

```python
model.fit(x_train, y_train)
```

```python
dfcoe = pd.DataFrame(model.coef_.flatten(), index=x.columns, columns=["Coefficient"])
print(dfcoe)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

### COMPARING ACTUAL TEST DATA AND PREDICTED DATA

```python
error= y_predicted - y_test
error
```

```python
pred = y_predicted.flatten() if hasattr(y_predicted, 'flatten') else y_predicted
y_true = y_test.values.flatten() if hasattr(y_test, 'values') else y_test
```

### MODEL EVALUATION

```python
sns.scatterplot(x=pred, y=y_true)
plt.xlabel("Predictions")
plt.ylabel("Yearly Amount Spent")
plt.title("Evaluation of our LM model")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
import math
```

```python
print("Mean Absolute Error: ",mean_absolute_error(y_test,y_predicted))
print("Mean Squared Error: ",mean_squared_error(y_test,y_predicted))
print("RMSE: ",math.sqrt(mean_squared_error(y_test,y_predicted)))
r2 = r2_score(y_test, y_predicted)
print("R² Score:", r2)
```
